```python
deprem = pd.read_excel("büyük_depremler.xls")
```

```python
deprem
```

```python
buyukluk_sınıf = []
for i in range(len(deprem)):
    if deprem['Büyüklük'][i] >= 8:
        buyukluk_sınıf.append('Great')
    elif (deprem['Büyüklük'][i] < 8) and (deprem['Büyüklük'][i] >= 7):
        buyukluk_sınıf.append('Major')
    elif (deprem['Büyüklük'][i] < 7) and (deprem['Büyüklük'][i] >= 6):
        buyukluk_sınıf.append('Strong')
    elif (deprem['Büyüklük'][i] < 6) and (deprem['Büyüklük'][i] >= 5):
        buyukluk_sınıf.append('Moderate')
    elif (deprem['Büyüklük'][i] < 5) and (deprem['Büyüklük'][i] >= 4):
        buyukluk_sınıf.append('Light')
    elif (deprem['Büyüklük'][i] < 4) and (deprem['Büyüklük'][i] >= 3):
        buyukluk_sınıf.append('Minor')
    else:
        buyukluk_sınıf.append('Very Minor')
deprem['buyukluk_sınıf'] = buyukluk_sınıf
```

```python
deprem.head()
```

```python
Açıklama = []
for i in range(len(deprem)):
    Açıklama.append('Tarih' + ':'  + str(deprem['Tarih'][i]) + '\n' + 'Buyukluk' + ':'  + str(deprem['Büyüklük'][i]) + "\n" + "Kayip" + ":" + str(deprem["Kayıplar"][i]))
    
deprem['Açıklama'] = Açıklama
```

```python
deprem.head()
```

```python
deprem.buyukluk_sınıf.unique()
```

```python
for i in range(len(deprem)):
    print(deprem["Enlem"][i], deprem["Boylam"][i])
```

```python
m = folium.Map(location = [39.1667, 35.6667], zoom_start=5)

for i in range(len(deprem)):
    if deprem['buyukluk_sınıf'][i] == 'Strong':
        folium.Marker(location= [deprem['Enlem'][i], deprem['Boylam'][i]], popup = deprem['Açıklama'][i], icon = folium.Icon(color = 'orange')).add_to(m)
    elif deprem['buyukluk_sınıf'][i] == 'Major':
        folium.Marker(location= [deprem['Enlem'][i], deprem['Boylam'][i]], popup = deprem['Açıklama'][i], icon = folium.Icon(color = 'red')).add_to(m)
    else:
        folium.Marker(location= [deprem['Enlem'][i], deprem['Boylam'][i]], popup = deprem['Açıklama'][i], icon = folium.Icon(color = 'yellow')).add_to(m)
    

m
```

```python
m = folium.Map(location=[39.1667, 35.6667], zoom_start=6)

choropleth = folium.Choropleth(geo_data = 'https://raw.githubusercontent.com/cads-tedu/DSPG/master/Veri%20Setleri/tr-cities-utf8.json',
                              data = deprem,
                              columns = ['İl', 'Büyüklük'],
                              key_on = 'properties.name',
                              fill_color = 'OrRd',
                              nan_fill_opacity=0,
                              fill_opacity = 0.7,
                              line_opacity = 0.5,
                              legend_name = 'Türkiyedeki En Büyük Depremler',
                              highlighted=True).add_to(m)

choropleth.geojson.add_child(folium.features.GeoJsonTooltip(['name'], labels = False))

m
```
